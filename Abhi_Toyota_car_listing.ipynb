# VISULIZATATION :-

# 1.Best Model In Car Listing

```python
ty['model'].value_counts()[:].plot(kind='line',color='r',figsize=(10,8),title='BEST MODEL').grid(color='black',linestyle='dotted',linewidth=2)
```

```python
plt.figure(figsize=(10,8))
ty.groupby('transmission')['model'].count().plot(kind='pie',autopct='%1.1f%%')
plt.legend(title='TRANSMISSION')
```

# 2.Checking Year Acccording EngineSize

```python
plt.subplots(figsize=(15,12))
sns.boxplot(x='year',y='engineSize',data=ty,palette='rainbow')
plt.show
```

```python
plt.figure(figsize=(15,11))
sns.violinplot(x='year',y='engineSize',data=ty)
plt.show
```

```python
ty.head(2)
```

# 3.Highest mpg used by toyota car listing

```python
plt.subplots(figsize=(20,7))
ty['mpg'].value_counts()[:].head(20).plot(kind='bar')
plt.title('HIGHEST MPG',color='r',fontsize=20)
plt.show()
```

# 4.Checking Which Transmission and Fueltype highest listing

```python
plt.subplots(figsize=(10,8))
ty['transmission'].value_counts()[:].plot(kind='bar',color='purple')
```

```python
plt.figure(figsize=(10,8))
ty['fuelType'].value_counts()[:].plot(kind='pie',autopct='%1.1f%%')
plt.legend(title='FUELTYPE')
```

```python
ty.head(3)
```

# 5.Highest one tax in car listing

```python
plt.figure(figsize=(15,5))
a = ty['tax'].value_counts()[:]
sns.barplot(x=a,y=a.index,palette='rainbow')
plt.title('No of Taxes')
plt.xlabel('How Many Times',size = 15)
plt.ylabel('Tax',size=15)
```

# 6.Price According Mileage

```python
plt.figure(figsize=(20,7))
sns.scatterplot(x='price',y='mileage',data=ty)
plt.xlabel('Price',size=20)
plt.ylabel('Mileage',size=20)
```

```python
sns.lmplot(x='price',y='mileage',data=ty)
plt.xlabel('Price',size=20)
plt.ylabel('Mileage',size=20)
```

```python
ty.head()
```

# 7.Which mpg is utilized by fueltype

```python
ty['fuelType']=ty['fuelType'].map({'Petrol':1,'Hybrid':2,'Diesel':3,'Other':4})
```

```python
plt.figure(figsize=(10,7))
sns.boxplot(x=ty['fuelType'],y=ty['mpg'])
plt.show
```
